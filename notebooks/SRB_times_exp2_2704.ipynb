```python
# analysis for one speed: -14
dfV = dfSRB2[dfSRB2['speed']==-14]
len(dfV)
```

```python
limmax = -6
limmin = -16
dfV['SRB'] = (dfV['U'] > limmin) & (dfV['U'] < limmax)
```

```python
print(len(dfV))
print(len(dfV[dfV['SRB']==True]))
```

```python
dfSRBtemp = dfV[dfV['SRB']==True]
dfSRBtemp['dtime'] = dfSRBtemp['Time'].diff(1)
df2 = pd.merge(dfV, dfSRBtemp[['dtime', 'Frame', 'session_id','Time']], on=['Frame', 'session_id','Time'], how='outer')

print(len(dfV))
print(len(df2))
df2.fillna(0, inplace=True)

dftemp = dfSRBtemp[(dfSRBtemp['dtime']>0.005) & (dfSRBtemp['dtime']<0.1)]
dftemp = dftemp.reset_index(drop=True)

maxF, minF = {}, {}
df2['SRBall'] =df2['SRB'] 

for i in np.arange(0,len(dftemp)):
# for i in np.arange(0,2):
    maxF[i] = dftemp.Time[i]
    minF[i] = df2[(df2['dtime']>0) & (df2['Time']<maxF[i])].Time
    minF[i] = minF[i].iloc[-1]
    print(i, minF[i], maxF[i])

    df2['SRBtt'] = np.where(((df2['Time']>minF[i]) & (df2['Time']<maxF[i])) , True, False)
    df2['SRBall'] = np.where((df2['SRBtt']==True) | (df2['SRBall']==True), True, False)
```

```python
a, b = 380, 400
g = sns.FacetGrid(data=df2, size=6, aspect=2, xlim=[a, b])
g.map(plt.plot, "Time", "SRB" , alpha=1, linewidth=1);

g = sns.FacetGrid(data=df2, size=6, aspect=2, xlim=[a, b])
g.map(plt.plot, "Time", "SRBall" , alpha=1, linewidth=1);


g = sns.FacetGrid(data=df2, size=6, aspect=2, xlim=[a, b])
g.map(plt.plot, "Time", "SRBall" , alpha=1, linewidth=1);
g.map(plt.plot, "Time", "UM" , alpha=1, linewidth=1, color='orange');
```

```python
# analysis for one speed: -14
dfV2 = dfSRB2[(dfSRB2['speed']==-14) | (dfSRB2['speed']==14)]
len(dfV2)
dfV2 = dfV2[(dfV2['speed']==-14) | (dfSRB2['speed']==14)]

dfV2['limmin'], dfV2['limmax'] = 0, 0
dfV2['limmin'] = np.where((dfV2['limmin']==0) & (dfV2['speed']==-14), -16, 0)
dfV2['limmax'] = np.where((dfV2['limmax']==0) & (dfV2['speed']==-14), -6 , 0)
dfV2['limmin'] = np.where((dfV2['limmin']==0) & (dfV2['speed']== 14),  6 , 0)
dfV2['limmax'] = np.where((dfV2['limmax']==0) & (dfV2['speed']== 14),  16, 0)
```
